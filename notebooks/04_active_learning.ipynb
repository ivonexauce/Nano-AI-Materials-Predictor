# 04 — Active Learning Experiment
Running the uncertainty-based active learning loop to discover new materials.

```python
import sys
sys.path.insert(0, '..')
from data.fetcher import generate_synthetic_samples
from prediction.active_learner import ActiveLearner

all_materials = generate_synthetic_samples(300)
labeled = all_materials[:50]
unlabeled = all_materials[50:]

learner = ActiveLearner(strategy='uncertainty')
learner.load_labeled_pool(labeled)
learner.load_unlabeled_pool(unlabeled)
history = learner.run_discovery_loop(n_rounds=3, query_size=10)
print(f'Total labeled: {len(learner.labeled_pool)}')
```
